```python
# 순위별로 프린트해보세요.
# 주로 많이 팔린 TOP10까지의 메뉴별 주문 개수
for i in range(len(item_count)):
    print('TOP ', i + 1, ':', item_count.index[i], '\t', item_count.values[i])
```

```python
# enumerate: 열거하다 
# Jquery의 $('list').each(): index랑 list의 항목을 하나씩 꺼내는 기능
# 시리즈의 values만 꺼내오고 싶으면, 시리즈를 enumerate에 입력값으로 넣어주면 됨.
# 시리즈를 tolist()하면 시리즈의 values만 리스트로 만들어진다.
for i, item in enumerate(item_count.iteritems(), 0): #index의 start값
    print('TOP ', i + 1, ':', item) #values만 꺼내옴
```

```python
for i, item in enumerate(item_count.iteritems(), 0): #index의 start값
    print('TOP ', i + 1, ':', item[0], item[1])
```

```python
for i, (index0, index1) in enumerate(item_count.iteritems(), 0): #index의 start값
    print('TOP ', i + 1, ':', index0, index1) # ?.. 이런 방법도 됨
```

```python
# enumerate 반복문: index, 반복할 item을 둘 다 꺼내주고 싶을 때 사용
data100 = [100, 200, 300]
for i, item in enumerate(data100, 0): #0은 index의 start값
    print(i, ':', item)
```

```python
result = []
for x in food['item_price']:
    result.append(float(x[1:]))
result[:10]
```

```python
food['item_price'] = result
```

# 과제 - 탐색적 분석

```python
# 1. 주문당 평균 계산금액 출력하기
food.groupby('order_id').mean()
```

```python
# 2. 한 주문에 10달러 이상 사용한 주문의 id들 출력하기
food[food['item_price'] >= 10]
```

```python
# 3. 각 아이템의 가격 구하기
food.groupby('item_name').min()
```

```python
# 4.가장 비싼 주문에서 item이 몇개 팔렸는지 구하기
food = food.sort_values(by='item_price' ,ascending=False)
food['quantity'].head(1)
```

```python
# 5."Veggie Salad Bowl"이 몇 번 주문되었는지 구하기
len(food[food['item_name'] == 'Veggie Salad Bowl'])
```

```python
# 6."Chicken Bowl"을 2개 이상 주문한 주문 횟수 구하기
food2 = food[food['item_name'] == 'Chicken Bowl']
len(food2[food2['quantity']>=2])
```
